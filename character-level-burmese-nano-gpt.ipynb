# there are some characters we don't need as u can see in outputs so we have to remove them


```python
unwanted_chars =  ['\u200b', '\u200c', '\u200d', '₠', '₡', '₢', '₣', '₤', '₥', '₦', '₧', '₨', '₩', '₪', '₫', '€', '₭', '₮', '₯', '₰', '₱', '₳', '₴', '₵', '₷', '₸', '₹', '₺', '₼', '₾']
for char in unwanted_chars:
    train_text = train_text.replace(char,'')
```

```python
chars = sorted(set(train_text))
vocab_size = len(chars)
print(f"vocab_size is {vocab_size}")
print(''.join(chars))
```

# character encoding and decoding
converting string to integer and integer to string

```python
stoi = {ch : i for i, ch in enumerate(chars)}
itos = {i : ch for i, ch in enumerate(chars)}
encode = lambda s: [stoi[c] for c in s]
decode = lambda l: ''.join([itos[int(i)] for i in l])
```

```python
print(encode('မင်္ဂလာပါ'))
print(decode(encode('မင်္ဂလာပါ')))
```

# enocde the entire dataset into torch.tensor

```python
import torch
data = torch.tensor(encode(train_text), dtype = torch.long)
print(data.shape, data[:1000])
```

# split the data into train and valid

```python
n = int(0.9 * len(data))
train_data = data[:n]
val_data = data[n:]
```

```python
block_size = 8
print(train_data[:block_size + 1])
print(decode(train_data[:block_size + 1]))
```

```python
x = train_data[0: block_size]#starts from 0
y = train_data[1:block_size+1]# y target always should be next token of input. so i starts from 0 and ends with block_size + 1
print(x,y)
```

```python
torch.manual_seed(2024) # for repreoducibillity
batch_size = 4 # for parallel processing using GPU
block_size =  8  #maximum context length

def get_batch(split):
    data = train_data if split == 'train' else val_data
    idx = torch.randint(len(data) - block_size, (batch_size,))#selecting random samples from data
    x = torch.stack([data[i:i+block_size] for i in idx])
    y = torch.stack([data[i+1:i+block_size+1] for i in idx])
    return x,y
```

```python
xb, yb = get_batch('train')
```

# building bigram language model which only looks the previous single word

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class BigramLanguageModel(nn.Module):
    def __init__(self, vocab_size):
        super().__init__()
        self.token_embedding_table = nn.Embedding(vocab_size, vocab_size)#vocab_size, embedding_dim
    
    def forward(self, idx, targets=None):
        logits = self.token_embedding_table(idx) #(B,T,C)/b = batch_size, t = number of characters or words in a sequence, c= number of features or word_embedding_dimension
        if targets is None:
            loss=None
        else:
            B,T,C = logits.shape
            logits = logits.view(B*T, C)
            targets = targets.view(B*T)
            loss = F.cross_entropy(logits, targets)
        
        return logits, loss
    
    def generate(self, idx, max_tokens):
        for _ in range(max_tokens):
            logits, loss = self(idx)
            logits = logits[:, -1, :]
            probs = F.softmax(logits, dim=1)
            idx_next = torch.multinomial(probs, num_samples=1)
            idx = torch.cat((idx, idx_next), dim=1)
        return idx
m = BigramLanguageModel(vocab_size)
logits, loss = m(xb,yb)
print(logits.shape)
print(loss)
print(decode(m.generate(idx = torch.zeros((1, 1), dtype=torch.long), max_tokens=100)[0].tolist()))
```

```python
optimizer = torch.optim.AdamW(m.parameters(), lr=1e-3)
```

# training

```python
batch_size=32
for steps in range(10000):
    xb, yb = get_batch('train')
    
    logits,loss = m(xb, yb)
    optimizer.zero_grad(set_to_none=True)
    loss.backward()
    optimizer.step()
    
    print(loss.item())
```
